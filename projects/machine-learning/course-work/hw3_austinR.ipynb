```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[0,0], [0,1], [1,0], [1,1]]) # Possible combinations
y = np.array([[0],[1],[1],[0]]) # Outcomes

# Model 1
model_1 = Sequential()
model_1.add(Dense(16, input_dim=2, activation='relu')) # Hidden layer w/ 16 neurons
model_1.add(Dense(1, activation='sigmoid'))
sgd = SGD(learning_rate=0.2) # Learning rate
model_1.compile(loss='mean_squared_error', optimizer=sgd) # Compiles model with mean squared error loss and sgd optimizer
model_1.fit(X, y, batch_size=1, epochs=100) # Fit the model running 100 epochs

rCount = 0
total = 0
for x in np.arange(0.0, 1.05, 0.05): # cannot use range since start and stop must be int
  for y in np.arange(0.0, 1.05, 0.05):
    if model_1.predict([[x,y]]) > 0.5:
      plt.plot(x,y,'r*') # to plot a fired point
      rCount += 1
      total += 1
    else: 
      plt.plot(x,y,'b.') # to plot a non-fired point
      total += 1
plt.show()

fPercent = (rCount / total) * 100 # Calculate fired percent

print('Fired Percent: %.2f' % fPercent)
```
